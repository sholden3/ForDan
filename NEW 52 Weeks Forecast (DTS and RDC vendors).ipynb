```python
#!pip install openpyxl==3.0.3
#!pip install xlrd==1.2.0

# import necessary packages
import pandas as pd
from openpyxl import load_workbook

# Like previously, run only one select statement through the workbook at a time
# Once you've created the files from one select statement, go back and run the next one all the way through


# RUN THIS ALWAYS - this is all non-DTS(Except BEHR)
sql = """SELECT * FROM `analytics-supplychain-thd.SXR5L14.RDC_DTS_COMBINED_MAIN`;"""


# RUN THIS TO GENERATE DTS FORECASTS FOR BEHR
#sql = """SELECT * FROM `analytics-supplychain-thd.SXR5L14.BEHR_COMBINED`; """

# read data from BQ
data = pd.read_gbq(sql, project_id = 'analytics-supplychain-thd', dialect = 'standard')
data
```

```python
# only run this line for non-DTS
#data = data.drop('indicator',axis=1)

# only run these lines for DTS
#data = data.drop('locId',axis=1)
#data['locId'] = 0

#only run this line for BEHR
data = data.drop('indicator',axis=1)

data
```

```python
# save just the pvendr column to a dataframe
pvendr = data['PVndr_nbr']

# function to get all unique pvendrs
def get_unique_numbers(pvendr):
    list_of_unique_numbers = []
    unique_numbers = set(pvendr)
    for number in unique_numbers:
        list_of_unique_numbers.append(number)
    return list_of_unique_numbers

# save unique pvendrs to a list
pvendr_unique = get_unique_numbers(pvendr)

# how many pvendr files do we have?
len(pvendr_unique)
```

```python
# loop through each pvendr to get just the data associated with it
for i in pvendr_unique:
    data_new = data[(data.PVndr_nbr == i)]
    
    # aggregate columns by SKU
    # take out DNO_Flag_Week1_Includes_BackLogged_demand if we decide to drop it
    data_agg = data_new.groupby(['PVndr_nbr','Pvndr_nm','Mvndr_nbr',
                'Mvndr_nm','SKU_NBR','SKU_DESC','department','class','Subclass','DNO_Flag_Week1_Includes_BackLogged_demand','Active_Vendor_Max_Parm','locId']).agg({'Avg_retl_amt':'mean',
                                                'week01_Current_week':'sum','week02':'sum','week03':'sum','week04':'sum','week05':'sum','week06':'sum',
                                                'week07':'sum','week08':'sum','week09':'sum','week10':'sum','week11':'sum','week12':'sum','week13':'sum',
                                                'week14':'sum','week15':'sum','week16':'sum','week17':'sum','week18':'sum','week19':'sum','week20':'sum',
                                                'week21':'sum','week22':'sum','week23':'sum','week24':'sum','week25':'sum','week26':'sum',
                                                'week27':'sum','week28':'sum','week29':'sum','week30':'sum','week31':'sum','week32':'sum',
                                                'week33':'sum','week34':'sum','week35':'sum','week36':'sum','week37':'sum','week38':'sum','week39':'sum',
                                                'week40':'sum','week41':'sum','week42':'sum','week43':'sum','week44':'sum','week45':'sum','week46':'sum',
                                                'week47':'sum','week48':'sum','week49':'sum','week50':'sum','week51':'sum','week52':'sum'}).reset_index()
    
    # read in the Legal Notice template
    legal_path = r"C:\Users\SWH2673\OneDrive - The Home Depot\Desktop\tests\LegalNotice.xlsx"
    legal = pd.read_excel(legal_path, header=None)
    
    # create the new filepath that we will save each vendor forecast to
    filepath = r"C:\Users\SWH2673\OneDrive - The Home Depot\Desktop\tests\(P" + str(i) + ".R2697)VendorSKUOrderForecast.xlsx"
    
    # load Legal Notice workbook and determine sheets within the file
    writer = pd.ExcelWriter(filepath, engine='openpyxl')
    book = load_workbook(legal_path)
    writer.book = book
    writer.sheets = dict((ws.title, ws) for ws in book.worksheets)

    # choose sheets and sheet names to write to new vendor files
    legal.to_excel(writer, sheet_name="Legal Notice", header=None, index=False)
    data_agg.to_excel(writer, sheet_name="Location Nation Summary", index=False)

    # save file
    writer.save()
```
